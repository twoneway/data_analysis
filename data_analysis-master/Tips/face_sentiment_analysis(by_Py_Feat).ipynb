```python
import pandas as pd
import numpy as np
```

```python
sentiment_list.to_csv("/content/drive/MyDrive/Colab Notebooks/sentiment analysis/output.csv")
raw_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/sentiment analysis/output.csv")
```

```python
en_data = raw_data[['anger', 'disgust', 'fear', 'happiness', 'sadness', 'surprise', 'neutral']]
en_data
```

```python
emotion_list = []
max_list = en_data.max(axis=1)

for a in range(len(en_data)):
  for b in range(len(en_data.columns)):
    if en_data.loc[a][b] == max_list[a] :
      emotion_list.append(en_data.loc[a].index[b])
```

```python
pro_data = raw_data[['FaceScore', 'Pitch', 'Yaw', 'anger', 'disgust', 'fear', 'happiness', 'sadness','surprise', 'neutral', 'input']]
pro_data['emotion'] = emotion_list
```

```python
label_list = []
for a in range(len(pro_data)):
  if 'unfocused' in pro_data['input'][a] :
    label_list.append(1)
  else : 
    label_list.append(0)
```

```python
pro_data['label'] = label_list
```

```python
for a in range(len(pro_data)) :
  pro_data['input'][a] = os.path.basename(pro_data['input'][a])
```

```python
fo = pro_data.iloc[:7]
ufo = pro_data.iloc[7:]
```

```python
pro_data['predict'] = 0

for a in range(len(pro_data)) :
  if pro_data['emotion'][a] != 'neutral':
    pro_data['predict'][a] = 1
  else :
    if (pro_data['Yaw'][a]>5.4 or pro_data['Yaw'][a]<-5.4 or pro_data['Pitch'][a]>8.66 or pro_data['Pitch'][a]<-8.66) :
      pro_data['predict'][a] = 1
    else : 
      pro_data['predict'][a] = 0      
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, precision_score, recall_score, plot_roc_curve, roc_curve, accuracy_score
```

```python
y_true = pro_data['label'].tolist()
y_pred = pro_data['predict'].tolist()
```

```python
print(confusion_matrix(y_true, y_pred))
print("AUC : ", roc_auc_score(y_true, y_pred))
print(classification_report(y_true, y_pred))
```

```python
# 다중 분류 모델의 confusion matrix를 그리는 함수 생성
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)
    plt.figure(figsize=(6.4,4.8))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]),   range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
        
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
import numpy as np
import itertools
```

```python
classes_list = ['focused', 'unfocused']
confusion_mtx = confusion_matrix(y_true, y_pred)
plot_confusion_matrix(confusion_mtx, classes=classes_list)
```

```python
print(confusion_matrix(y_true, y_pred))
print(roc_auc_score(y_true, y_pred))
print(precision_score(y_true, y_pred))
print(recall_score(y_true, y_pred))
print(accuracy_score(y_true, y_pred))
print(classification_report(y_true, y_pred))
```

```python
def plot_roc_curve(fper, tper):
    plt.plot(fper, tper, color='red', label='ROC')
    plt.plot([0, 1], [0, 1], color='green', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic Curve')
    plt.legend()
    plt.show()
```

```python
fper, tper, thresholds = roc_curve(y_true, y_pred)
plot_roc_curve(fper, tper)
```
